```python
import arxiv
import requests
import json
import csv
import pandas as pd
from collections import Counter, defaultdict
import numpy as np # for array manipulation
import matplotlib.pyplot as plt # for data visualization
%matplotlib inline 
import datetime
```

```python
search = arxiv.Search(
  query = "cat:math.CO",
  max_results = 6,
  sort_by = arxiv.SortCriterion.SubmittedDate
)

for result in search.results():
  print(result.title, result.entry_id, result.published)
```

```python
results_generator = arxiv.Client(
  page_size=1000,
  delay_seconds=3,
  num_retries=3
).results(arxiv.Search(
  query="cat:math.CO",
  id_list=[],
  sort_by=arxiv.SortCriterion.Relevance,
  sort_order=arxiv.SortOrder.Descending,
))fetched

CO_papers = []
for paper in results_generator:
  # You could do per-paper analysis here; for now, just collect them in a list.
  CO_papers.append(paper)
```

```python
co_df = pd.DataFrame([vars(paper) for paper in CO_papers])
```

```python
co_df.to_csv('43k_CO_papers.csv')
```
